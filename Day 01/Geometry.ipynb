### **Question 2**

Can you write code to find distance of the line $x = 50$ and $y = 25$ from the origin?

Can you derive the formula to compute distance between two parallel lines from the formulas provided in the previous section?



```
import numpy as np

def distanceOrigin(a, b, c):
    return abs(c)/np.sqrt(a**2 + b**2)

def distanceOriginSmart(a, b, c):
  w = [a, b]
  # linalg.norm(w) is nothing but np.sqrt(w[0]**2 + w[1]**2)
  return np.absolute(c) / np.linalg.norm(w)
  

# Line x = 50
a = 1
b = 0
c = -50
print(distanceOrigin(a, b ,c))

# Line y = 25
a = 0
b = 1
c = -25
print(distanceOriginSmart(a, b ,c))
```

```
def distParallels(a1, b1, c1, a2, b2, c2):
  return abs(c1 - c2)/np.sqrt(a1**2 + b1**2)

def distParallelsSmart(a1, b1, c1, a2, b2, c2):
  w = [a1, b1]
  return abs(c1 - c2)/np.linalg.norm(w)

# Line y = 25
a1 = 0
b1 = 1
c1 = -25
# Line y = 5
a2 = 0
b2 = 1
c2 = -5
print(distParallels(a1, b1, c1, a2, b2, c2))
print(distParallelsSmart(a1, b1, c1, a2, b2, c2))
```

### **Challenge Question:**
Can you write the relation between tanh and sigmoid functions?

```
def sigmoid(X):
    return 1/(1 + np.exp(-X))

def tanh(X):
    return (np.exp(X) - np.exp(-X))/(np.exp(X) + np.exp(-X))

def gaussian(X):
    return np.exp(-np.square(X))

def reLU(X):
    boolX = (X > 0).astype(int)
    return X*boolX

def leaky_reLU(X):
    bool1 = (X > 0.1*X).astype(int)
    bool2 = (X <= 0.1*X).astype(int)
    
    return X*bool1 + bool2*0.1*X

def step(X):
    bool1 = (X > 0).astype(int)
    return bool1 + 0*X
```

The hyperbolic tangent ($tanh$) and logistic sigmoid ($\sigma$) functions are defined as follows:
\begin{equation}
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
\;\;\;\; \sigma(x) = \frac{1}{1 + e^{-x}}
\end{equation}

```
X = np.linspace(-5, 5, num = 200)
plt.plot(X, sigmoid(X), 'm-', label = 'sigmoid')
plt.plot(X, tanh(X), 'b-', label = 'tanH')
plt.title("Sigmoid and tanh")
plt.legend()
```


Sigmoid $\sigma(x)$ being an odd function can be written as,

\begin{equation}
1 - \sigma(x) =  \sigma(-x)
\end{equation}

That is,

\begin{equation}
 1 -\frac{1}{1 + e^{-x}} = \frac{1}{1 + e^{x}}
\end{equation}

Now,
\begin{equation}
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
\end{equation}

\begin{equation}
        = \frac{e^x + e^{-x} - 2e^{-x}}{e^x + e^{-x}} 
\end{equation}

\begin{equation}
        = 1+  \frac{ - 2e^{-x}}{e^x + e^{-x}}  
\end{equation}

\begin{equation}
        = 1 - \frac{2e^{-x}}{e^{2x} + 1} 
\end{equation}

Now substituting the sigmoid function we get,

\begin{equation}
tanh(x) = 1 - \frac{2e^{-x}}{e^{2x} + 1} 
\end{equation}

\begin{equation}
        = 1 - 2\sigma(-2x) 
\end{equation}

\begin{equation}
        = 1 - 2 + 2\sigma(2x)
\end{equation}

\begin{equation} 
        = 2\sigma(2x)-1
\end{equation}



Plotting the above we get,

```
X = np.linspace(-10, 10, num = 200)
plt.plot(X, 2*sigmoid(2*X) - 1, label = 'tanH')
plt.title("tanH Using Sigmoid")
plt.legend()
```

### Understanding *np.logspace()* and *np.linspace()*

The logspace function generates logarithmically spaced vectors. Especially useful for creating frequency vectors, it is a logarithmic equivalent of linspace and the ":" or colon operator.

$y = np.logspace(a,b,n)$ generates $n$ points between decades $10^a$ and $10^b$. Default $n$ is 50.

The linspace generating linearly spaced values.

$y = np.linspace(a,b,n)$ generates $n$ points between decades $a$ and $b$. Default $n$ is 100.


### Plotting two dimenaional shapes using linspace

### ***Circle***

The equation is,

\begin{equation}
x^2 + y^2 = R^2
\end{equation}

```
r = 5
x = np.linspace(-r, r, num = 200)
Yplus = np.sqrt(r * r - x * x)
Yminus = -np.sqrt(r * r - x * x)
plt.xlim(-20, 20)
plt.ylim(-15, 15)
plt.title("Circle")
plt.plot(x, Yplus, 'steelblue', x, Yminus, 'gold', label = 'Radius: ' + str(r))
plt.legend(loc = 'upper left')
```

### ***Ellipse***

The equation is,

\begin{equation}
\frac{x^2}{a^2}+\frac{y^2}{b^2} = 1
\end{equation}

```
a = 2
b = 4
X = np.linspace(-a, a, num = 100)
Yplus = b*np.sqrt(1 - (X**2/a**2))
Yminus = -Yplus
plt.plot(X, Yplus, 'c-', X, Yminus, 'g-' , label = 'a: ' + str(a) + ' b: ' + str(b))
plt.title("Ellipse")
plt.legend(loc = 'upper left')
```

### ***Square***


```
a = 5

X = np.linspace(0, a, num = 200)
Yplus = 0*X
Yminus = Yplus + a
plt.title("Square")
plt.xlim(-10, 15)
plt.ylim(-6, 10)
plt.plot(X, Yplus, X, Yminus, label = 'a: ' + str(a))
plt.plot(Yplus, X, Yminus, X, label = 'a: ' + str(a))
plt.legend(loc = 'upper left')
```

### ***Contour Plots***

Here the 2D Gaussian function has been visualized:
\begin{equation}
f(x) = e^{-x^2}
\end{equation}

The purpose of **np.meshgrid()** is to create a rectangular grid out of an arra of $x$ values and an array of $y$ values.

Choosing Colour Maps! https://matplotlib.org/3.2.1/tutorials/colors/colormaps.html

```
%matplotlib inline
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
f = np.exp(-X*X - Y*Y)
C = plt.contour(X, Y, f)
plt.set_cmap('hot') # Used to change the colour map!
plt.title('Gaussian Contour Plot')
plt.clabel(C)
```

### Three dimensional plots!



### ***Plane***

The equation of a plane is
\begin{equation}
ax + by + cz + d = 0
\end{equation}

```
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
# Setting up parameters
a = 1
b = 1
c = 3
d = 10
xx = np.linspace(-10, 10, 200)
yy = np.linspace(-10, 10, 200)
X,Y  = np.meshgrid(xx, yy)
Z = (-a*X - b*Y - d) * 1.0 / c
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, Y, Z, cmap='coolwarm')
plt3d
```
